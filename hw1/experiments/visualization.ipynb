## 1.3 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df = pd.read_csv('1.3_results.csv')
```

```python
grad_steps = df['num_agent_train_steps_per_iter']
avg_return = df['Average Return']
std_return = df['Std Return']
```

```python
rcParams['figure.figsize'] = 8, 5
sns.set()
ax = sns.lineplot(x='num_agent_train_steps_per_iter', y='Average Return', data=df)
upper_bound = avg_return + std_return
lower_bound = avg_return - std_return 
ax.set(xlabel='Gradient Step Size', title='Gradient Step Size v.s. Average Return on Ant-v2')
ax.fill_between(grad_steps, upper_bound, lower_bound, alpha=.3)
ax.figure.savefig("1.3_visualization.png")
```

## 2.2 Results

```python
df = pd.read_csv('2.2_results.csv')
```

```python
rcParams['figure.figsize'] = 8, 5
sns.set()
ax = sns.lineplot(x='Epoch', y='Eval_AverageReturn', data=df, label="DAgger")
avg_return = df['Eval_AverageReturn']
std_return = df['Eval_StdReturn']
upper_bound = avg_return + std_return
lower_bound = avg_return - std_return 
ax.set(xlabel='Epoch', ylabel="Average Return", title='Epoch v.s. Average Return on Ant-v2')
ax.fill_between(df['Epoch'], upper_bound, lower_bound, alpha=.3)
ax.axhline(4713.65, ls='--', c='red', label="Expert")
ax.axhline(4090.54, ls='--', c='green', label="Behavior Cloning")
plt.legend()
ax.figure.savefig("2.2_visualization.png")
```
